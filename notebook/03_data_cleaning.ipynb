```python
def clean_column_names(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("-", "_")
    )
    return df
```

```python
specs_df = clean_column_names(specs_df)

benchmarks_df = clean_column_names(benchmarks_df)

api_df = clean_column_names(api_df)
```
