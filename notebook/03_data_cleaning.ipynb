```python
def clean_column_names(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("-", "_")
    )
    return df
```

```python
specs_df = clean_column_names(specs_df)

benchmarks_df = clean_column_names(benchmarks_df)

api_df = clean_column_names(api_df)
```

```python
specs_df["gpu_name"] = specs_df["gpu_name"].str.strip()

benchmarks_df["gpu_name"] = benchmarks_df["gpu_name"].str.strip()

api_df["gpu_name"] = api_df["gpu_name"].str.strip()
```

```python
specs_df["gpu_name"] = specs_df["gpu_name"].str.upper()

benchmarks_df["gpu_name"] = benchmarks_df["gpu_name"].str.upper()

api_df["gpu_name"] = api_df["gpu_name"].str.upper()
```

```python
common_specs_bench = set(specs_df["gpu_name"]).intersection(
    set(benchmarks_df["gpu_name"])
)

print("Common GPUs:", len(common_specs_bench))
```

```python
common_bench_api = set(benchmarks_df["gpu_name"]).intersection(
    set(api_df["gpu_name"])
)

print("Common GPUs:", len(common_bench_api))
```

```python
master_df = pd.merge(
    specs_df,
    benchmarks_df,
    on="gpu_name",
    how="inner"
)
```

```python
master_df.drop(
    columns=["manufacturer_y"],
    inplace=True
)

master_df.rename(
    columns={"manufacturer_x":"manufacturer"},
    inplace=True
)
```

```python
missing_percent = (
    master_df.isnull().sum()
    / len(master_df)
) * 100

missing_percent.sort_values(
    ascending=False
)
```

```python
msno.matrix(master_df)
```

```python
columns_to_drop = [
    "membuswidth",
    "metal",
    "cuda",
    "vertexshader",
    "pixelshader",
    "vulkan"
]

master_df.drop(
    columns=columns_to_drop,
    inplace=True
)
```

```python
master_df.drop_duplicates(inplace=True)
```

```python
master_df["memsize"] = master_df["memsize"].fillna(
    master_df["memsize"].median()
)

master_df["memclock"] = master_df["memclock"].fillna(
    master_df["memclock"].median()
)

master_df["unifiedshader"] = master_df["unifiedshader"].fillna(
    master_df["unifiedshader"].median()
)
```

```python
master_df["memtype"] = master_df["memtype"].fillna(
    master_df["memtype"].mode()[0]
)
```
