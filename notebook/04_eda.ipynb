```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
df = pd.read_csv(
    "../data/cleaned/cleaned_gpu_dataset.csv"
)

df.head()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["memsize"],
    bins=30,
    kde=True
)

plt.title("GPU Memory Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["gpuclock"],
    bins=30,
    kde=True
)

plt.title("GPU Clock Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["g3dmark"],
    bins=30,
    kde=True
)

plt.title("G3D Benchmark Distribution")
plt.show()
```

```python
numerical_cols = [
    "memsize",
    "gpuclock",
    "memclock",
    "g3dmark",
    "g2dmark"
]

for col in numerical_cols:

    print(f"\n{col}")

    print("Mean:", df[col].mean())

    print("Median:", df[col].median())

    print("Mode:", df[col].mode()[0])

for col in numerical_cols:

    print(f"\n{col}")

    print("Variance:", df[col].var())

    print("Std Dev:", df[col].std())


for col in numerical_cols:

    print(f"\n{col}")

    print("Skewness:", df[col].skew())

    print("Kurtosis:", df[col].kurt())
```

```python
numeric_df = df.select_dtypes(
    include=np.number
)

corr_matrix = numeric_df.corr()
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix")
plt.show()
```

```python
corr_matrix["g3dmark"].sort_values(
    ascending=False
)
```

```python
df["performance_per_watt"] = (
    df["g3dmark"] / df["tdp"]
)

df["performance_per_gb"] = (
    df["g3dmark"] / df["memsize"]
)

df["performance_per_shader"] = (
    df["g3dmark"] / df["unifiedshader"]
)

df["gpu_age"] = (
    2026 - df["releaseyear"]
)

df["performance_category"] = pd.qcut(
    df["g3dmark"],
    q=4,
    labels=[
        "Low",
        "Mid",
        "High",
        "Extreme"
    ]
)
```
