```python
import pandas as pd
import numpy as np

from scipy import stats

df = pd.read_csv(
    "../data/processed/feature_engineered_gpu_dataset.csv"
)
```

```python
mean = df["g3dmark"].mean()

std = df["g3dmark"].std()

n = len(df)

margin = 1.96 * (std / np.sqrt(n))

lower = mean - margin
upper = mean + margin

print("Mean:", mean)

print("95% Confidence Interval:")

print(lower, upper)
```

```python
new_gpu = df[
    df["releaseyear"] >= 2015
]["g3dmark"]

old_gpu = df[
    df["releaseyear"] < 2015
]["g3dmark"]

t_stat, p_value = stats.ttest_ind(
    new_gpu,
    old_gpu,
    equal_var=False
)

print("T Statistic:", t_stat)

print("P Value:", p_value)
```

```python
amd = df[
    df["manufacturer"] == "AMD"
]["g3dmark"]

nvidia = df[
    df["manufacturer"] == "NVIDIA"
]["g3dmark"]

ati = df[
    df["manufacturer"] == "ATI"
]["g3dmark"]


f_stat, p_value = stats.f_oneway(
    amd,
    nvidia,
    ati
)

print("F Statistic:", f_stat)

print("P Value:", p_value)
```

```python
corr = df.corr(
    numeric_only=True
)

corr["g3dmark"].sort_values(
    ascending=False
)
```
